# Make figures that use data from all 3 ciphers

```python
import numpy as np
import tensorflow as tf
import os

%matplotlib inline
import matplotlib.pyplot as plt
```

### Load generalization data for each cipher task

```python
ciphers = ['autokey', 'vigenere', 'enigma']
cipher_gens = []
for c in ciphers:
    with open('saved_np_arrays/generalize-{}.txt'.format(c), 'r') as f:
        g = np.loadtxt(f, delimiter=',', unpack=True).T
        cipher_gens += [g]
```

### No generalization (control example)
If the model performed perfectly for the first 20 time steps that it was trained on and then did not generalize AT ALL, this is what it's generalization curve would look like

```python
k = 5
msg_len_hist = []
max_msg_len = 200
msg_lens = list(range(10,max_msg_len,k))

for msg_len in msg_lens:
    acc = min(100*20/msg_len, 100.0)
    msg_len_hist += [[msg_len, acc]]
    
mlh = np.vstack(msg_len_hist)
    
ciphers += ['no generalization']
cipher_gens += [mlh]
```

```python
f = plt.figure(figsize=[6,4], dpi=100)
Ciphers = [c[0].upper() + c[1:] for c in ciphers]
styles = ['-', '--', '-.', ':']

for i, g in enumerate(cipher_gens):
    plt.plot(g[:,0], g[:,1], linewidth=3.0, label=Ciphers[i], ls=styles[i])
plt.xlabel('message length', fontsize=14) ; plt.setp(plt.gca().axes.get_xticklabels(), fontsize=14)
plt.ylabel('accuracy (%)', fontsize=14) ; plt.setp(plt.gca().axes.get_yticklabels(), fontsize=14)
plt.ylim([50,101])
plt.legend()

plt.show(f) #; f.savefig('generalize-all.png', bbox_inches='tight')
```
